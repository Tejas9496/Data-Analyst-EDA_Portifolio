```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel("flight_price.xlsx")
```

```python
df.sample(20)
```

```python
df['Date_of_Journey'].unique()
```

```python
#Feauture Engineering
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df['Date'].unique()
```

```python
#type casting
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
#Drop  Date of  journey Column
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Arrival_Time'].unique()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.sample(5)
```

```python
df.drop('Arrival_Time',inplace=True,axis=1)
```

```python
df.info()
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_min']=df['Dep_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time',inplace=True,axis=1)
```

```python
df.sample(5)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, np.nan:1, '4 stops':4})
```

```python
df['Total_Stops'].unique()
```

```python
df.drop('Route',inplace=True,axis=1)
```

```python
df['Duration_hrs']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df['Duration_mins']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```
